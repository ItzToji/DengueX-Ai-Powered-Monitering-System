```python
%%writefile /kaggle/working/dengue.yaml
path: /kaggle/input/mosquito/yolo

train: images/train
val: images/val
test: images/test

names:
  0: dengue_mosquito
  1: non_dengue_mosquito
  2: non_mosquito
```

```python
from ultralytics import YOLO

# 🔑 Load the checkpoint you want to resume from
model = YOLO("/kaggle/working/runs/dengue_yolo_v1/weights/last.pt")

model.train(
    resume=True,                     # 🔑 REQUIRED
    data="/kaggle/working/dengue.yaml",
    epochs=22,                       # total target epochs
    imgsz=640,
    batch=32,
    device=0,
    workers=2,
    patience=10,
    project="/kaggle/working/runs",
    name="dengue_yolo_v1",            # MUST be same name
    save=True,
    save_period=1
)
```
